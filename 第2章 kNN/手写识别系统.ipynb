```python
import numpy as np
import os
import operator
```

```python
def classify0(inx,d,l,k):
    ds = d.shape[0]
    diffMat = np.tile(inx,(ds,1))-d
    distance = (diffMat**2).sum(axis=1)**0.5
    # print(distance)
    sortedDistIndicies = distance.argsort()
    classCount = {}
    for i in range(k):
        voteLabel = l[sortedDistIndicies[i]]
        classCount[voteLabel] = classCount.get(voteLabel,0)+1
    sortedClassCount = sorted(classCount.items(),key=operator.itemgetter(1),reverse=True)
    # print(classCount)
    return sortedClassCount[0][0]
```

```python
def image2vector(filename):
    returnVect = np.zeros((1,1024))
    fr = open(filename)
    for i in range(32):
        lineStr = fr.readline()
        for j in range(32):
            returnVect[0,32*i+j] = int(lineStr[j])
    return returnVect
```

```python
os.listdir('digits/testDigits')
```

```python
def hwCT():
    hwLabels = []
    trainingFileList = os.listdir('digits/trainingDigits')
    m = len(trainingFileList)
    trainingMat = np.zeros((m,1024))
    for i in range(m):
        filename = trainingFileList[i]
        fileStr = filename.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        hwLabels.append(classNumStr)
        trainingMat[i,:] = image2vector('digits/trainingDigits/%s'%filename)
    testFileList = os.listdir('digits/testDigits')
    errorCount = 0
    mTest = len(testFileList)
    for i in range(mTest):
        filename = testFileList[i]
        fileStr = filename.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        vectorUnderTest = image2vector("digits/testDigits/%s"%filename)
        classifierResult = classify0(vectorUnderTest,trainingMat,hwLabels,3)
        print("predict:%d\ntrue:%d"%(classifierResult,classNumStr))
        if classifierResult != classNumStr:
            errorCount += 1
    print("error rate:%f"%(errorCount/mTest))
```

```python
hwCT()
```
